```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ds_salaries.csv').sample(frac = 1)
df[:][:-1]
```

```python
df.drop(['Unnamed: 0','salary', 'salary_currency','employee_residence','company_location'], inplace=True, axis=1)
df.head()
```

```python
df.info()
```

```python
df_num = (df.remote_ratio - df.remote_ratio.min()) / (df.remote_ratio.max() - df.remote_ratio.min())
df.drop('remote_ratio', inplace=True, axis=1)
df
```

```python
df_Y = pd.DataFrame(df['salary_in_usd'])
df_Y.rename(columns = {'salary_in_usd': 'Salary(USD)'}, inplace = True)
print(f'Y : \n{df_Y}')
df_X = df[df.columns.drop('salary_in_usd')]
print(f'X : \n{df_X}')
```

```python
enc = OneHotEncoder()
```

```python
enc.fit(df_X)
dfX = enc.transform(df_X).toarray()
```

```python
df_dX = pd.DataFrame(dfX)
df_dX
```

```python
df_concat = pd.concat([df_dX, df_num], axis = 1)
```

```python
df_concat
```

```python
df_concat.to_csv('prepared_data.csv')
df_Y.to_csv('pd_y.csv')
```
